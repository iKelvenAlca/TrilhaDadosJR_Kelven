```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Total de compras por categoria
df['Total de Vendas'] = df['Quantidade de Vendas'] * df['Preço Unitário']
df_grouped = df.groupby('Nome do Curso', as_index=False)['Total de Vendas'].sum()
print(df_grouped)
```

```python
#Grafico Total vendido por curso
plt.bar(df_grouped['Nome do Curso'], df_grouped['Total de Vendas'], color='red')
plt.xlabel('Nome do Curso')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Curso')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

O curso mais introdutório da programação, HTML e CSS, foi o mais vendido. Em último temos o curso de segurança da informação.

```python
#Curso mais vendido
mais_vendido=df_grouped['Total de Vendas'].max()
print(df_grouped.loc[df_grouped['Total de Vendas'] == mais_vendido])
```

```python
#Distribuição de vendas por tempo
data=df['Data']
vendas=df['Total de Vendas']
plt.plot(data, vendas, color='red')
plt.xlabel('Data')
plt.ylabel('Total de Vendas')
plt.title('Distribuição de Vendas por Tempo')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Excelente distribuição de vendas para um único mês, sendo que em nenhum dia não houve vendas.
